```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',header=None)
```

```python
header = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peek-rpm","city-mgp","highway-mpg","price"]
df.columns=header
df.head()
```

```python
df1=df.replace('?',np.NaN)
```

```python
df=df1.dropna(subset=['price'],axis=0)
```

```python
#identify count of missing values

missing_data=df.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
   
```

```python
#Deal with missing data - drop all the rows and column
missing_data.dropna()
```

```python

#Deal with missing data - drop all the rows

df.dropna()
```

```python
#Deal with missing data - drop all columns with missing value 

df.dropna(axis=1)
```

```python
#Deal with missing data - drop all rows only if all the values in the row is missing,
#use how='any' to drop all the rows if atleast one of the value is missing 
df.dropna(how='all')
```

```python
#Deal with missing data - replace the missing value with mean

mean_normalized_losses = df["normalized-losses"].astype("float").mean(axis=0)
mean_normalized_losses
df["normalized-losses"].replace(np.nan,mean_normalized_losses,inplace=True)
df
```

```python
#Deal with missing data - replace the missing value with mean(cloumn bore)
mean_bore=df["bore"].astype("float").mean(axis=0)
mean_bore
df["bore"].replace(np.nan,mean_bore,inplace=True)
df
```

```python
#Deal with missing data - replace the missing value with mean for datatype as string

#1. find the most repeated strng value

df["num-of-doors"].value_counts().idxmax

#2. replace with the most repeated value

df["num-of-doors"].replace(np.nan,"four")
df
```

```python
# enquire data type
df.dtypes
```

```python
# replace the data type to right format
#normalized-losses ,bore,stroke,price,peek-rpm
df[["normalized-losses"]] =df [["normalized-losses"]].astype("int")
df[["bore","stroke"]] = df[["bore","stroke"]].astype("float")
df[["peek-rpm"]] = df[["peek-rpm"]].astype("float")
df[["price"]] = df[["price"]].astype("float")
df.dtypes
```

```python
#Data normalisation - dividing with maximum value(obtain values betwwen 0-1)

df['length']= df['length']/df['length'].max()
df['length'] .values
```
